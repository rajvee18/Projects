## Data Preprocessing

```python
df = pd.read_csv('/home/code_wizard/coding_files/Projects/Media and Technology/Global YouTube Statistics.csv', encoding='latin-1')

df
```

```python
df.isnull().sum()
```

```python


df['subscribers'] = df['subscribers'].fillna(df['subscribers'].mean())

df.isnull().sum()
```

#### Q1: top 10 youtube cahnnels based on subscribers

```python
top_10_sub = df.sort_values(by='subscribers',ascending=False)

top_10_sub.head(10)
```

### 2.	Which category has the highest average number of subscribers?

```python
highest_average_subscribers_category = df.groupby(['category'])['uploads'].mean()
print(highest_average_subscribers_category.sort_values(ascending=False).head(1))
```

### 3.	How many videos, on average, are uploaded by YouTube channels in each category?

```python
average_upload_by_category = df.groupby(['category'])['uploads'].mean()

print(average_upload_by_category.sort_values(ascending=False))
```

### 4.	What are the top 5 countries with the highest number of YouTube channels?

```python
country_with_more_channels = df['Abbreviation'].value_counts().head(5)

print(country_with_more_channels)

```
