## Importing the required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Uploading the dataset

```python
df = pd.read_csv('/home/code_wizard/coding_files/Projects/Media and Technology/Global YouTube Statistics.csv', encoding='latin-1')

df
```

## Data Cleaning

```python
df.isnull().sum()
```

```python


df['subscribers'] = df['subscribers'].fillna(df['subscribers'].mean())
df['created_year'] = df['created_year'].fillna(df['created_year'].mean())
df['Unemployment rate'] = df['Unemployment rate'].fillna(df['Unemployment rate'].mean())
df['Urban_population'] = df['Urban_population'].fillna(df['Urban_population'].mean())

df['Latitude'] = df['Latitude'].fillna(df['Latitude'].median())
df['Longitude'] = df['Longitude'].fillna(df['Longitude'].mean()) 

df['subscribers_for_last_30_days'] = df['subscribers_for_last_30_days'].fillna(df['subscribers_for_last_30_days'].median())
df['Gross tertiary education enrollment (%)'] = df['Gross tertiary education enrollment (%)'].fillna(df['Gross tertiary education enrollment (%)'].mean())

df.isnull().sum()
```

### 12.	How does the unemployment rate vary among the top 10 countries with the highest number of YouTube channels?

```python



country_channel_counts = df['Country'].value_counts()

# Step 2: Identify the top 10 countries with the highest number of YouTube channels
top_10_countries = country_channel_counts.head(10).index

# Step 3: Filter the DataFrame to include only the top 10 countries
top_10_df = df[df['Country'].isin(top_10_countries)]

# Step 4: Aggregate the unemployment rate data for each of these top 10 countries
unemployment_data = top_10_df.groupby('Country')['Unemployment rate'].mean()

# Step 5: Visualize the data
plt.figure(figsize=(14, 7))
unemployment_data.sort_values().plot(kind='bar')
plt.title('Average Unemployment Rate Among the Top 10 Countries with the Highest Number of YouTube Channels')
plt.xlabel('Country')
plt.ylabel('Average Unemployment Rate')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### 15.	What is the correlation between the number of subscribers and the population of a country?

```python


country_data = df.groupby('Country').agg({
    'subscribers': 'sum',
    'Population': 'mean'  # Assuming population is constant for each country
}).reset_index()

# Calculate the correlation between the number of subscribers and the population
correlation = country_data['subscribers'].corr(country_data['Population'])

print("Correlation between the number of subscribers and the population of a country:", correlation)
```

### 16.	How do the top 10 countries with the highest number of YouTube channels compare in terms of their total population?

```python
country_channel_counts = df['Country'].value_counts()

# Step 2: Identify the top 10 countries with the highest number of YouTube channels
top_10_countries = country_channel_counts.head(10).index

# Step 3: Filter the DataFrame to include only the top 10 countries
top_10_df = df[df['Country'].isin(top_10_countries)]

# Step 4: Aggregate the population data for each of these top 10 countries
population_data = top_10_df.groupby('Country')['Population'].mean()

# Step 5: Visualize the data
plt.figure(figsize=(12, 6))
population_data.sort_values().plot(kind='bar')
plt.title('Total Population of the Top 10 Countries with the Highest Number of YouTube Channels')
plt.xlabel('Country')
plt.ylabel('Total Population')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### 17.	Is there a correlation between the number of subscribers gained in the last 30 days and the unemployment rate in a country?

### 20.	What is the average number of subscribers gained per month since the creation of YouTube channels till now?

```python
df['created_date'] = pd.to_datetime(df['created_date'])

# Calculate the current date
current_date = pd.to_datetime('now')

# Calculate the number of months each channel has been active
df['Months_Active'] = (current_date.year - df['created_date'].dt.year) * 12 + (current_date.month - df['created_date'].dt.month)

# Calculate the average number of subscribers gained per month for each channel
df['Subscribers_per_Month'] = df['subscribers'] / df['Months_Active']

# Calculate the overall average number of subscribers gained per month
average_subscribers_per_month = df['Subscribers_per_Month'].mean()

print("Average number of subscribers gained per month since the creation of YouTube channels:", average_subscribers_per_month)
```
